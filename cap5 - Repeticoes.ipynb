```python
# 5.14
contador = 0
montante = 0
while True:
    valor = int(input("Digite um valor ou 0 para sair: "))
    if valor == 0:
        break
    montante = montante + valor
    contador = contador + 1
    media = montante/contador

print(contador)
print(media)
    
    
```

```python
# 5.15

valor_final = 0
quantTotal = 0
valor = 0
preço = 0
quantidade = 0
while True:  
    codigo = int(input("Digite o codigo relacionado ao produto: "))
    if (codigo == 0):
        break
    elif (codigo == 1):
        preço = 0.5
    elif (codigo == 2):
        preço = 1
    elif (codigo == 3):
        preço = 4
    elif (codigo == 5):
        preço = 7
    elif (codigo == 9):
        preço = 8
    else:
        print("Código invalido")
        break
        
    quantidade = int(input("Digite a quantidade do produto selecionado: "))

    valor = preço * quantidade
    valor_final += valor
    quantTotal += quantidade

print(f"Preço: R${valor_final}")
print(f"Quantidade: {quantTotal}")
```

```python
# 5.16

valor = float(input("Digite o valor a pagar:"))

cedulas = 0
atual = 100
apagar = valor

while True:
    if atual <= apagar:
        apagar -= atual
        cedulas += 1
    else:
        print(f"{cedulas} cédula(s) de R${atual}")
        if apagar == 0:
            break
        if atual == 100:
            atual = 50
        elif atual == 50:
            atual = 20
        elif atual == 20:
            atual = 10
        elif atual == 10:
            atual = 5
        elif atual == 5:
            atual = 1
        elif atual == 1:
            atual = 0.50
        elif atual == 0.50:
            atual = 0.10
        elif atual == 0.10:
            atual = 0.05
        elif atual == 0.05:
            atual = 0.02
        elif atual == 0.02:
            atual = 0.01
        elif atual == 0.01:
            atual = 0.001
        cedulas = 0
```

### Repetições aninhadas

```python
tabuada = 1

while tabuada <= 10:
    numero = 1
    while numero <= 10:
        print(f"{tabuada} x {numero} = {tabuada * numero}")
        numero += 1
    tabuada += 1
```

```python
# 5.22
opcao = 0
tabuada = 1
while (opcao != 5):
    opcao = int(input("Escolha uma opção:\n 1-Adição \n 2-Subtração \n 3-Divisão \n 4-Multiplicação \n 5-Sair \n Resposta: "))
    if (opcao == 1):
        while tabuada <=10:
            numero = 1
            while numero <= 10:
                print(f"{tabuada} + {numero} = {tabuada + numero}")
                numero += 1
            tabuada +=1
    if (opcao == 2):
        while tabuada <=10:
            numero = 1
            while numero <= 10:
                print(f"{tabuada} - {numero} = {tabuada - numero}")
                numero += 1
            tabuada +=1
    if (opcao == 3):
        while tabuada <=10:
            numero = 1
            while numero <= 10:
                print(f"{tabuada} / {numero} = {tabuada / numero}")
                numero += 1
            tabuada +=1
    if (opcao == 4):
        while tabuada <=10:
            numero = 1
            while numero <= 10:
                print(f"{tabuada} x {numero} = {tabuada * numero}")
                numero += 1
            tabuada +=1
    tabuada = 1
    numero = 1
    
        
```
